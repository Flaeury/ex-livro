```python
import math
largura_blocos=32
cache= 128 * 2**10
#como a linha informada esta em hexadecimal, e a quantidade de algarismos, quando transformada em binario, é multiplicada por 4 ja que 16 = 2^4
adress = 7*4
campo_byte = math.log2(largura_blocos)
quantidade_de_linhas = cache / largura_blocos
campo_linhas=math.log2(quantidade_de_linhas)
campo_tag= adress-campo_linhas-campo_byte

#11111111001 0010010101 10110 

print ("O tamanho do campo de tags é :", campo_tag)
print ("O tamanho do campo de linhas é :", campo_linhas)
print ("O tamanho do campo de bytes é :", campo_byte)
```

Os bits mais significativos serao atribuidos na tag (11111111001)
Os bits subsequentes serão atribuidos à linha (0010010101)
Os menos sgnificativos irão para o byte (10110)
